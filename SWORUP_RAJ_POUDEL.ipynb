```python
import numpy as np                          #mathematical computation library
import matplotlib.pyplot as plt             #visualization library
```

**SOLVING QUESTION NO 9**

*   mass(m) in kg
*   stiffness(k) in N/m
*   displacement(x) from the equilibrium position in meter
*   velocity(v)  in meter/sec
*   damping constant(c) 
*   damp frequency(Wd) in rad/sec
*   natural frequency(Wn) in rad/sec
*   critical damping(Cc)


```python
#variable assignment
m= 10  
k= 1000
x_0= 0.05
v_0= 0.5
```

```python
def zeta_cond(zeta):
  if zeta>1:
    print('overdamp case'.upper())
  elif zeta==1:
    print('critical damp case'.upper())
  elif zeta<1:
    print('underdamp case'.upper())
```

```python
displacement=[]
t= np.arange(0,5,0.01)
def ploting_graph(zeta,Wn, Wd):
  if zeta < 1:
    p = -zeta*Wn*t
    a = (v_0+zeta*Wn*x_0)/Wd
    b = x_0
    theta = Wd*t
    x_t = np.exp(p)*(a*np.sin(theta)+b*np.cos(theta))
    x_t= x_t.real

    

  elif zeta == 1:
    x_t = np.exp(-Wn*t)*(x_0 + (v_0+Wn*x_0)*t)






  elif zeta > 1:
    a = (v_0+(zeta+np.sqrt(zeta*zeta-1))*Wn*x_0)/(2*np.sqrt(zeta*zeta-1)*Wn)
    b = -(v_0+(zeta-np.sqrt(zeta*zeta-1))*Wn*x_0)/(2*np.sqrt(zeta*zeta-1)*Wn)
    p = (-zeta+np.sqrt(zeta*zeta-1))*Wn*t
    q = (-zeta-np.sqrt(zeta*zeta-1))*Wn*t
    x_t = a*np.exp(p)+b*np.exp(q)
  displacement.append(x_t)
  plt.plot(t, x_t)
  plt.xlabel('time(sec)')
  plt.ylabel('displacement(m)')
  plt.show()
```

```python
def find_zeta_wn_wd(c,m=10,k=1000,x_0=0.05, v_0=0.5):
  Wn = np.sqrt(k/m)
  Cc= 2* np.sqrt(k*m)
  zeta= c/Cc
  Wd= Wn*np.sqrt(1-np.power(zeta, 2))
  zeta_cond(zeta)
  print('============================')
  print('Natural Frequecy: {}'.format(Wn))
  print('Critical damping: {}'.format(Cc))
  print('Zeta value: {}'.format(zeta))
  print('Damping frequency: {}'.format(Wd))
  print('============================')
  ploting_graph(zeta,Wn, Wd)
```

When damping constant is 50 N.s/m

```python
find_zeta_wn_wd(50)
```

When damping constant is 200 N.s/m

```python
find_zeta_wn_wd(200)
```

When damping constant is 300 N.s/m

```python
find_zeta_wn_wd(300)
```
